## Objective
Test varying gRNA concentration with both DNAse and non-DNAse treated gRNA, test with and without pre-assembly

## Formatting to CSV file
Take txt file from plate reader and turn into tidy data format

```python
#import needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
#read in CSV file
DETECTR_raw = pd.read_csv("../raw_txt_files/DETECTR_006.txt", encoding='utf-16', sep="\t", delimiter="\t",index_col=None, skiprows= 3,header = None)
```

```python
#drop empty columns - we only had samples in the 2 and 3 columns of the plate
DETECTR_raw = DETECTR_raw.drop(list(range(11,len(DETECTR_raw.columns))), axis=1)
DETECTR_raw = DETECTR_raw.drop(list(range(1,8)), axis=1)
```

```python
DETECTR_raw.head(10)
```

```python
#identify the end of the file
end_indices = DETECTR_raw[0][DETECTR_raw[0]=='~End'].index.tolist()
end_indices[0]
```

```python
#remove unnecessary colums at the beginning of the file and unnecessary rows at the end of the file
#we want to keep the file with the time information
DETECTR_raw = DETECTR_raw.drop(list(range(end_indices[0], len(DETECTR_raw))), axis = 0)
```

```python
#how many plate reads did we have?
len(DETECTR_raw)/16
```

```python
#set a variable to identify the number of plate reads
num_plate_reads = int(len(DETECTR_raw)/16)
```

```python
DETECTR_raw.head(16)
```

```python
#create a dictionary mapping the row number to a letter for 384 and 96
well_dict_384 = {1:"A",2:"B",3:"C",4:"D",5:"E",6:"F",7:"G",8:"H",9:"I",10:"J",11:"K",12:"L",13:"M",14:"N",15:"O",16:"P"}
well_dict_96 = {1:"A",2:"A",3:"B",4:"B",5:"C",6:"C",7:"D",8:"D",9:"E",10:"E",11:"F",12:"F",13:"G",14:"G",15:"H",16:"H"}
```

```python
#initialize the cleaned data frame
DETECTR_cleaned = pd.DataFrame(columns = ['time','row_384','column_384', 'row_96','column_96','rfu_value'])
```

```python
#iterate over each plate to create a dataframe with the correct time, the 384 and 96 position and the RFU_value
for i in list(range(0,num_plate_reads)):
    time = DETECTR_raw.iloc[i*16,0]
    for j in list(range(8, 11)):
        for k in list(range(i*16, i*16+16)):
            rfu_value = DETECTR_raw.loc[k,j]
            row_384 = well_dict_384[k%16+1]
            column_384 = j - 1
            row_96 = well_dict_96[k%16+1]
            column_96 = j // 2 
            DETECTR_cleaned.loc[len(DETECTR_cleaned)] = [time, row_384, column_384,row_96,column_96,rfu_value]
    
```

```python
#write this file to a CSV
DETECTR_cleaned.to_csv("DETECTR_006_tidy.csv")
```

## Filter the data frame
Select only wells we are interested in

```python
#only select rows with the 2 hour (final) time point
DETECTR_4hr = DETECTR_cleaned.loc[(DETECTR_cleaned["time"]=="04:00:00") ]
```

```python
#only select the rows in which we had sample
DETECTR_4hr = DETECTR_4hr.loc[(DETECTR_4hr["row_384"].isin(['A','C','E','G','I','K','M','O']))]
DETECTR_4hr = DETECTR_4hr.drop(DETECTR_4hr.loc[(DETECTR_4hr["row_384"].isin(['I','K','M','O'])) & DETECTR_4hr["column_384"].isin([7])].index, axis = 0)
```

```python
DETECTR_4hr
```

```python
#define the experimental time which target DNA and gDNA was added
DETECTR_4hr['Setup'] = pd.Series(["Pre-assembly","Pre-assembly","Pre-assembly","Pre-assembly","non-DNAse-treated","non-DNAse-treated","non-DNAse-treated","non-DNAse-treated","non-DNAse-treated","non-DNAse-treated","non-DNAse-treated","non-DNAse-treated","DNAse-treated","DNAse-treated","DNAse-treated","DNAse-treated","DNAse-treated","DNAse-treated","DNAse-treated","DNAse-treated"], index = DETECTR_4hr.index)
```

```python
#define the gRNA + target DNA used for this set
DETECTR_4hr['gRNA'] = pd.Series(["12A","12A","12A","12A","12C","12C","12C","12C","12C","12C","12C","12C","12C","12C","12C","12C","12C","12C","12C","12C"], index = DETECTR_4hr.index)
```

```python
#define whether gDNA was added
DETECTR_4hr['gRNA concentration'] = pd.Series([100, 100, 100,100, 800,400,200,100,50,25,12.5,0,800,400,200,100,50,25,12.5,0], index = DETECTR_4hr.index)
```

```python
DETECTR_4hr['well'] = (DETECTR_4hr['row_384'].map(str) + DETECTR_4hr['column_384'].map(str)).map(str)
```

## Plot the data
Create plots for each experiment set

```python
DETECTR_4hr['rfu_value'] = DETECTR_4hr['rfu_value'].map(float)
DETECTR_4hr['gRNA concentration'] = DETECTR_4hr['gRNA concentration'].map(float)
```

```python
background = statistics.mean(DETECTR_4hr.loc[(DETECTR_4hr["column_384"].isin([8,9])) & (DETECTR_4hr["row_384"].isin(['O']) )]['rfu_value'])
background
```

```python
DETECTR_4hr['rfu_value_bkgrnd_sub'] = DETECTR_4hr['rfu_value'] - background
```

```python
DETECTR_4hr.loc[DETECTR_4hr['Setup'] == 'Pre-assembly'].plot(x = 'well', y= 'rfu_value_bkgrnd_sub',kind = 'bar')
```

```python
DETECTR_4hr.loc[DETECTR_4hr['Setup'].isin(['DNAse-treated','non-DNAse-treated'])]
```

```python
DETECTR_4hr.loc[DETECTR_4hr['Setup'].isin(['DNAse-treated','non-DNAse-treated'])].pivot('gRNA concentration', 'Setup','rfu_value_bkgrnd_sub').plot(kind = 'line')
plt.title ('DETECTR_006 (Time = 4hr)')
plt.ylabel('RFU_value - background')
plt.rc('font', size=10)
plt.rc('axes', labelsize=15)

plt.savefig("DETECTR_006.pdf", bbox_inches='tight')
```

## Conclusions
The gRNA concentration significantly increases the signal of the RFU_value. It would appear that DNAse treating the guides has helped remove ssDNA template which may be interfering with the reaction. However, at a certain point, the more gRNAs, the lower the signal. As cas12a has some not insignifigant trans activity on ssRNA, it is possible that by flooding the reaction with gRNAs which do not complex with the cas12a can detract from cleavage of the ssDNA reporter and reduce signal.
